# Eigenwertzerlegung

"Primfaktorzerlegung für Matrizen": $12 = 2 \times 2 \times 3$

LGS: $A \vec{v} = \lambda \vec{v}$

$\vec{v}$: (rechter) <b>Eigenvektor</b> zu A

$\lambda$: <b>Eigenwert</b> zum Eigenvektor $\vec{v}$

<b>Eigenwertzerlegung</b>: $A = Qdiag(\lambda)Q^{-1}$

$Q$: <b>orthogonale Matrix</b> ($Q^{-1} = Q^T$): Vektoren sind <b>orthonormal</b>: sie stehen senkrecht zueinander ($x^Ty=0$) und haben die Norm 1

# Count-based Methoden

## One-Hot-Encoding

Vokabular: {Ich, mag, Äpfel, aber, keine, Birnen}

Ich = $\begin{pmatrix} 1 \\ 0 \\ 0 \\ 0 \\ 0 \\ 0 \end{pmatrix}$,
Äpfel = $\begin{pmatrix} 0 \\ 0 \\ 1 \\ 0 \\ 0 \\ 0 \end{pmatrix}$,
Birnen = $\begin{pmatrix} 0 \\ 0 \\ 0 \\ 0 \\ 0 \\ 1 \end{pmatrix}$


```python
from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer()
corpus = [
          'The cat sat on the hat',
          'The dog ate the cat and the hat',
          ]
# Erstelle Matrix mit Sätzen in Zeile und Wörtern in Spalten
# Zähle, wie oft jedes Wort in jedem Satz vorkommt
X = vectorizer.fit_transform(corpus)
print(vectorizer.get_feature_names())

# Embedding Matrix
print(X.toarray())
# Embedding für den Satz "A new document"
vectorizer.transform(['The cat ate the mouse']).toarray()
# 'a','new' nicht Teil des Vokabulars
```

## TF-IDF Embeddings

tf-idf =  term frequency * inverse document frequency

tf-idf misst die Wichtigkeit eines Wortes in einem Korpus. Das Gewicht eines Wortes steigt, je öfter es im Dokument vorkommt, aber wird mit der Worthäufigkeit im Korpus gegengerechnet

<b>Worthäufigkeit (TF)</b>: $TF(t) = \frac{\text{# Wort t im Dokument}}{\text{# Wörter im Dokument}}$

<b>Inverse Dokumenthäufigkeit (IDF)</b>: $IDF(t) = log_e(\frac{\text{# Dokumente}}{\text{# Dokumente mit Wort t}})$

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer()
corpus = [
          'The cat sat on the hat',
          'The dog ate the cat and the hat',
]
# Erstelle Matrix mit Sätzen in Zeile und Wörtern in Spalten
# Zähle, wie oft jedes Wort in jedem Satz vorkommt
X = vectorizer.fit_transform(corpus)
print(vectorizer.get_feature_names())

# Embedding Matrix
print(X.toarray())
# Embedding für den Satz "A new document"
vectorizer.transform(['The cat ate the mouse']).toarray()
```

<b>Nachteile dieser Methoden</b>:

1) Keine numerische Ähnlichkeit zwischen semantisch ähnlichen Wörtern (Äpfel $\approx$ Birnen)

2) Matrix enthält viele Nullen (<b>sparse matrix</b>)

## Positive Pointwise Mutual Information (PPMI)

Treten Wörter 1 und 2 zusammen häufiger auf als unabhängig voneinander? 

$$PPMI(word_1, word_2) = max(log_2 \frac{P(word_1, word_2)}{P(word_1)P(word_2)},0)$$

PPMI tendiert seltenen Wörtern hohe Werte zu geben

1) Erhöhe die Wahrscheinlichkeit für seltene Wörter

2) Nutze Add-One Smoothing
