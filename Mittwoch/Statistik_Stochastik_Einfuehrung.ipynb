# N-Gramm-Modelle

> Ich mag Äpfel, aber keine Birnen.

> $w_1 w_2 w_3 w_4 w_5 w_6 w_7 w_8$

gesucht: $$P(w_1,\dots, w_8)$$

Idee: Die Wahrscheinlichkeit des <b><i>i-ten</i></b> Wortes wird in Abhängigkeit vom Kontext aus den <b><i>i-1</i></b> vorhergehenden Wörtern berechnet:

$$ P(w_i|w_1,\dots,w_{i-1}) $$

<b>Bigramm-Modell</b>: $N = 2$, Wort an Position $i$ hängt nur von Wort an Position $i-1$ ab

$P(w_i | w_1, \dots, w_{i-1}) = P(w_i | w_{i-1})$

<b>Trigramm-Modell</b>: $N = 3$, Wort an Position $i$ hängt nur von Wörtern an Positionen $i-1$ und $i-2$ ab

$P(w_i | w_1, \dots, w_{i-1}) = P(w_i | w_{i-1}, w_{i-2})$

# Wahrscheinlichkeitsfunktionen, -dichten und verteilungen?

<b>diskrete Zufallsvariable (discrete random variable)</b>:  beschreibt die <b>Wahrscheinlichkeitsfunktion (probability mass function)</b> $P(X = x_i) = P(x_i) = p_i$

Es gilt:

1) Definitionsmenge von Abbildung $P$: alle möglichen Zustände der Zufallsvariable $X$, z.B. $Q=$\{Birnen, Bananen, Trauben\}

2) $\forall x_i \in Q : 0 \leq P(x_i) \leq 1$

3) $\sum_{x_i \in Q} P(x_i) = 1$ (Normalisierung)

Beispiel: <b> diskrete Gleichverteilung (uniform distribution)</b>: 

\begin{equation*}
P(X=x_i) = f(x) = \begin{cases}
\frac{1}{k} &\text{für $x = x_i, (i = 1, \dots, k)$}\\
0 &\text{sonst}
\end{cases}
\end{equation*}

# Wahrscheinlichkeitsfunktionen, -dichten und verteilungen?

<b>stetige Zufallsvariable (continuous random variable)</b>: wie häufig der jeweilige Wert der Zufallsvariable angenommen wird, beschreibt die <b>Wahrscheinlichkeitsdichte (probability density function)</b>

Es gilt:

1) Definitionsmenge von Abbildung $P$: alle möglichen Zustände der Zufallsvariable $X$

2) $\forall x_i \in x: p(x_i) \geq 0$

3) $\int_{x_i \in x} p(x_i) dx_i = 1 $

Die Wahrscheinlichkeitsdichte beschreibt nicht direkt die Wahrscheinlichkeit eines bestimmten Zustands, sondern die Wahrscheinlichkeit in einer unendlich kleinen Region um diesen Zustand zu landen. 

Beispiel: <b> stetige Gleichverteilung </b>:

\begin{equation*}
f(x) = \begin{cases}
\frac{1}{b-a} &\text{für $ a \leq x \leq b$}\\
0 &\text{sonst}
\end{cases}
\end{equation*}

<b>Wahrscheinlichkeitsverteilung (probability distribution)</b>

diskreter Fall: zu der Wahrscheinlichkeitsfuntkion gehört eine Verteilungsfunktion $$F(x)=P(X\leq x)$$ als Summe aller Wahrscheinlichkeiten der Ausprägungen kleiner oder gleich x gegeben. Diese Wahrscheinlichkeit wird auch als <b>kumulierte oder summierte Wahrscheinlichkeit</b> bezeichnet.

stetiger Fall: im Falle einer stetigen Wahrscheinlichkeitsverteilung erhältst du die zugehörige Verteilungsfunktion durch Integration der Dichtefunktion 
$$F(x)=\int_{-\infty}^x f(t) dt $$

# Weitere wichtige Begriffe

<b>Randverteilung (marginal probability distribution)</b>: Wahrscheinlichkeitsverteilung einer Teilmenge einer Zufallsvariablenmenge

Beispiel: Zufallsvariablen X und Y, Wahrscheinlichkeitsverteilung von X = Randverteilung der Verteilung (X, Y)

Stichwort: Kontingenztafel



<b>Erwartungswert</b>: Mittelwert der Ergebnisse bei mehrmaligem Wiederholen eines Experiments

diskret: $E(X) = \sum_x x \cdot P(X = x)$

stetig: $E(X) = \int_a^b x \cdot f(x) dx $ für $x \in [a,b]$

Es gilt Linearität: $E(aX + bY) = a \cdot E(X) + b \cdot E(Y)$

<b>Varianz</b>: wie sehr variieren die Werte einer Funktion einer Zufallsvariable, wenn man verschiedene Werte aus der Verteilung zieht

$V(X) = E((X-E(X))^2)$

<b>Standardabweichung</b>: $\sqrt{V(X)}$

<b>Bedingte Wahrscheinlichkeit (conditional probability)</b>: Wahrscheinlichkeit, dass ein Ereignis Y eintritt, nachdem das Ereignis X eingetreten ist

$$P(Y = y | X = x) =\frac{P(Y = y, X = x)}{P(X = x)}$$

<b>Satz von Bayes</b>: 

$$P(Y = y | X = x) = \frac{P(X = x | Y = y)P(Y = y)}{P(X = x)}$$

### Kettenregel bedingter Wahrscheinlichkeiten

$$P(x_1, \dots, x_n) = P(x_1) \prod_{i=2}^{n} P(x_i | x_1, \dots, x_{i-1}) $$

Beispielssatz: $$P(w_1, \dots, w_8) = P(w_1) \prod_{i=2}^8 P(w_i | w_1, \dots, w_{i-1})$$

### Unabhängigkeit und bedingte Unabhängigkeit

zwei Zufallsvariablen X, Y sind <b>unabhängig</b> ($X \perp Y$), wenn ihre Wahrscheinlichkeitsverteilung als Produkt zweier Faktoren ausgedrückt werden kann, die nur von jeweils einer Zufallsvariable abhängen:

$$ \forall x \in X, y \in Y: P(X=x, Y=y) = P(X = x) * P(Y = y)$$

zwei Zufallsvariablen X, Y sind <b>bedingt unabhängig gegeben Z</b> ($X \perp Y |Z $), wenn ihre Wahrscheinlichkeitsverteilung als Produkt zweier Faktoren ausgedrückt werden kann, die von z und jeweils einer Zufallsvariable abhängen:
    
$$ \forall x \in X, y \in Y, z \in Z: P(X=x, Y=y| Z=z) = P(X=x|Z=z)*P(Y=y|Z=z) $$

> Ich mag Äpfel, aber keine Birnen.

> $w_1 w_2 w_3 w_4 w_5 w_6 w_7 w_8$

gesucht: $$P(w_1,\dots, w_8)$$

Annahme: einzelne Wörter sind stochastisch unabhängig

$P(w_1, \dots, w_8) = P(w_1) \prod_{i=2}^8 P(w_i | w_1, \dots, w_{i-1})=\prod_{i=1}^8 P(w_i)$

Macht diese Annahme Sinn?

gesucht: Schätzer für $P(w_i)$

Methode: <b>Maximum-Likelihood-Schätzung</b>: die Wahrscheinlichkeit eines Wortes ist gleich der relativen Häufigkeit im Trainingscorpus

$P(w_7 = \text{"Birnen"}) = \frac{\text{#"Birnen"}}{\text{#Wörter im Text}}$

## Stochastischer Prozess

<b>stochastischer Prozess</b>: Folge von Zufallsereignissen $X_1, X_2, \dots X_n$

Die möglichen Zufallswerte in einem stochastischen Prozess heißen <b>Zustände des Prozesses</b>. Man sagt, dass sich der Prozess zum Zeitpunkt $t$ in Zustand $X_t$ befindet.

Hier: Zeitpunkt $\approx$ Wortposition, Zustand $\approx$ Wort

charakterisiert durch <b>Anfangswahrscheinlichkeit</b> und <b>Übergangswahrscheinlichkeiten</b>

## Markow-Kette (spezieller stochatischer Prozess)

Die Wahrscheinlichkeit des <b><i>i-ten</i></b> Wortes im Kontext von <b><i>i-1</i></b> vorhergehenden Wörtern kann angenähert werden mit der Wahrscheinlichkeit des <b><i>i-ten</i></b> Wortes im verkürzten Kontext der vorhergehenden <b><i>k-1</i></b> Wörter (<b>Markow-Kette k-ter Ordnung</b>)

$P(x_1, \dots, x_n) = P(x_1) \prod_{i=2}^n P(x_i | x_1, \dots, x_{i-1})$ 

$= P(x_1) \prod_{i=2}^n P(x_i | x_{i-(k-1)} , \dots, x_{i-1})$

<b>Markow-Kette</b>: "Die Zukunft bedingt auf der Gegenwart hängt nicht von der Vergangenheit ab" (<b>Gedächtnislosigkeit</b>): $P(x_{i+1}| x_1, \dots, x_{i}) = P(x_{i+1} | x_i)$ 

entspricht dem <b>Bigramm-Sprachmodell</b>


## Markow-Modell

Bei einem Markow-Modell wird jedem Zustand eine <b>Ausgabe</b> zugeordnet, die ausschließlich vom aktuellen <b>Zustand</b> abhängt.

Ausgabe: Sequenz von Ereignissen, die die Beobachtungen in der Beobachtungssequenz repräsentieren

### Hidden Markow Modell (HMM)

- nur die Sequenz der Ausgaben beobachtbar ist
- die Sequenz der Zustände verborgen bleibt

Es kann mehrere Zustandssequenzen geben, die dieselbe Ausgabe erzeugen


### Beispiel



## Hidden Markow Modell

$(S, K, \Pi, A, B)$

- $S = \{S_1, \dots, S_n\}$: Menge der Zustände
- $K = \{k_1, \dots, k_m\}$: Menge der Ausgabesymbole
- $\Pi = \{\pi\}, \pi = P(X_1 = S_i)$: Anfangswahrscheinlichkeiten
- $A = \{a_{ij}\}, a_{ij} = P(X_{t+1} = S_j | X_t = S_i)$: Übergangswahrscheinlichkeit der Zustände
- $B = \{b_j(k)\}, b_j(k) = P(K_k \quad \text{in} \quad t | X_t = S_j)$: Emissionswahrscheinlichkeit in Zustand j

Neben der Markow-Annahme gibt es noch die <b>Unabhängigkeit der Ausgabe</b>: Ausgabe in Zeitpunkt $t$ hängt nur von Zustand $X_t$ ab und nicht von vorherigen Ausgaben

    - Forward/Backward-Algorithmus
- Beste Pfadsequenz finden
    - Viterbi-Algorithmus
- Training: Aufbau des besten Modell aus Trainingsdaten

# Generative und diskriminative Modelle

Klassisisches Thema in NLP: sequentielle Zuordnung von Labels ($y_i$) zu Beobachtungen ($x_i$) (= Klassifizierung)

Beispiel: Part-of-Speech-Tagging: Zuordnung der Wortart zu einem Wort

Daten: Paare aus Wörtern und Labels

gesucht: $P(y_1, y_2, \dots, y_n | x_1,x_2, \dots x_n)$



<b>generatives Modell</b>: separates Modell für jedes Label oder jede Klasse basierend auf Trainingsdaten für jede Klasse. Gegeben einer Beobachtung werden Vorhersagen für alle Klassen berechnet und wahrscheinlichste ausgewählt. Beispiele: Hidden Markow Modell, Naive Bayes Klassifikator

<b>diskriminatives Modell</b>: lernt Features (Merkmalsvektoren) von den Trainingsdaten, um zwischen Labels/Klassen zu unterscheiden. Beispiel: Logistische Regression

<b>Naive Bayes Klassifikator</b>: eine Klasse $y \in Y$ für eine Sequenz an Wörtern $x$

$$ y' = \text{argmax}_y p(y | x ) $$

Bayes' Regel: $y' = \text{argmax}_y \frac{P(x|y)P(y)}{P(x)} = \text{argmax}_y P(x|y) P(y)$

$P(y)$: apriori Wahrscheinlichkeit der Klasse $y$

$P(x|y)$: Wahrscheinlichkeit des Textes $x$ gegeben der Klasse $y$





<b>Naive Bayes Annahme</b>: wir können die Wortsequenz als Menge aus Merkmalsvektoren $f_1, \dots, f_n$ darstellen und nehmen an, dass die Merkmalsvektoren innerhalb einer Klasse unabhängig sind

$P(x |y) = P(f_1, \dots, f_n | y) = P(f_1|y) \dots P(f_n | y) $

jedes Wort ist ein Merkmal: 

$ y' = \text{argmax}_y P(y) \prod_{i=1}^{|x|} P(w_i|y) $

Die Wahrscheinlichkeiten werden aus den Trainingsdaten geschtzt

<b>Logistische Regression</b>

$$ p(y|x) = \frac{1}{Z} \text{exp}(\sum_{i=1}^N m_i f_i) $$

$Z$: Normalisierungsfaktor
$m_i$: Gewichte für Merkmalsvektoren $f_i$

## Maximum Entropy Markow Modell

verbindet sequentielle Zuordnung gegeben einer Sequenz aus Beobachtungen (HMM) mit multinomialer Logistischer Regression (auch genannt <b>Maximum Entropy</b>)

Übergangs- und Emissionswahrscheinlichkeiten werden zusammengefasst: $P(s_t | s_{t_1}, o_t)$

MEMM nutzt zusätzlich Merkmalsvektoren (feature vectors) und verallgemeinert die Übergangswahrscheinlichkeiten zu beliebigen Wörter im Input

MEMM modelliert weit entfernte Verbindungen in der Beobachtungssequenz zusammen mit dem vorhergehenden Zustand dank gemeinsamer Verteilung
